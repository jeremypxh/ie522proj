```python
import numpy as np

###############################################################################
# 1) All alpha/beta rate functions from the paper’s Table I (x1..x25 plugged in)
###############################################################################
def alpha111(V):
    return 3.802 / (
        0.1027*np.exp(-(V+2.5)/17.0)
      + 0.20  *np.exp(-(V+2.5)/150.0)
    )

def beta111(V):
    return 0.1917*np.exp(-(V+2.5)/20.3)

def alpha31(V):
    # x11=7, x23=7.7 => 7e-7 * exp(-(V+7)/7.7)
    return 7.0e-7*np.exp(-(V+7.0)/7.7)

def beta31(V):
    # x12=0.0084 + 2e-5*(V+7)
    return 0.0084 + 2.0e-5*(V+7.0)

def alpha112(V):
    # same as alpha12 but for the IC2->IF arrow
    return 3.802 / (
        0.1027*np.exp(-(V+2.5)/15.0)
      + 0.23  *np.exp(-(V+2.5)/150.0)
    )

def beta112(V):
    # same as beta12 for IF->IC2
    return 0.20*np.exp(-(V+2.5)/20.3)

def alpha4(V):
    # arrow IF->I1
    # x18=0.001 => alpha4 = x18 * alpha2, but in the diagram it’s just labeled “alpha4”.
    # We'll define a direct function. If the paper says alpha4 is literally "0.001 * alpha2(V)", do that:
    # Or if the paper’s Table I has alpha4 separate, plug that in.
    # For brevity, let’s do alpha4= x18*a2 => 0.001 * alpha2(V):
    return 0.001*alpha2(V)

def beta4(V):
    # arrow I1->IF
    # x19=0.02 => beta4 = x19 * alpha33 => 0.02*alpha33(V)
    return 0.02*alpha33(V)

def alpha5(V):
    # arrow I1->I2
    # x20=1/95000 => alpha5= x20* alpha2 => alpha2(V)/95000
    return alpha2(V)/95000.0

def beta5(V):
    # arrow I2->I1
    # x21=0.02 => beta5= x21* alpha33 => 0.02*alpha33(V)
    return 0.02*alpha33(V)

def alpha11(V):
    # arrow C2->IC2
    return alpha111(V)  # The paper lumps alpha11 and alpha111 the same formula
def beta11(V):
    # arrow IC2->C2
    return beta111(V)

def alpha12(V):
    # arrow C2->C1
    return 3.802 / (
        0.1027*np.exp(-(V+2.5)/15.0)
      + 0.23  *np.exp(-(V+2.5)/150.0)
    )
def beta12(V):
    # arrow C1->C2
    return 0.20*np.exp(-(V+2.5)/20.3)

def alpha13(V):
    # arrow C1->O
    return 3.802 / (
        0.1027*np.exp(-(V+2.5)/12.0)
      + 0.25  *np.exp(-(V+2.5)/150.0)
    )
def beta13(V):
    # arrow O->C1
    return 0.22*np.exp(-(V+2.5)/20.3)

def alpha33(V):
    return alpha31(V)  # often the same expression used for alpha31.. alpha33

def beta33(V):
    return beta31(V)

def alpha2(V):
    # arrow IF->O
    # x17=7, x22=16.6 => alpha2=0.188495* exp(-(V+7)/16.6 + 0.393956)
    return 0.188495*np.exp(-(V+7.0)/16.6 + 0.393956)

def beta2(V):
    # arrow O->IF
    # from the figure: labeled "beta33 * beta2"
    # the code snippet: beta2 = alpha13*alpha2*alpha33/(beta13*beta33)
    # We'll define it as that:
    return alpha13(V)*alpha2(V)*alpha33(V)/(beta13(V)*beta33(V))


###############################################################################
# 2) BUILD THE COMPLETE 9×9 MATRIX A(V), EXACTLY MATCHING FIG. 6 ARROWS
###############################################################################
def build_A_matrix(V):
    """
    Indices (1-based):
      1: IC3
      2: IC2
      3: IF
      4: I1
      5: I2
      6: C3
      7: C2
      8: C1
      9: O
    A_{i,j} = rate from j->i.
    Diagonal: negative sum of rates leaving j.
    """

    A = np.zeros((9,9))

    #-- IC3 <-> IC2: alpha111 from IC3->IC2, beta111 from IC2->IC3
    #   i=2, j=1 => A[1,0] = alpha111
    A[1,0] = alpha111(V)      # IC3->IC2
    A[0,0] -= alpha111(V)     # out of state #1

    #   i=1, j=2 => A[0,1] = beta111
    A[0,1] = beta111(V)       # IC2->IC3
    A[1,1] -= beta111(V)      # out of state #2

    #-- IC3 <-> C3: alpha31 from C3->IC3, beta31 from IC3->C3
    #   i=1, j=6 => A[0,5] = alpha31
    A[0,5] = alpha31(V)       # C3->IC3
    A[5,5] -= alpha31(V)

    #   i=6, j=1 => A[5,0] = beta31
    A[5,0] = beta31(V)        # IC3->C3
    A[0,0] -= beta31(V)

    #-- IC2 <-> IF: alpha112 from IC2->IF, beta112 from IF->IC2
    #   i=3, j=2 => A[2,1] = alpha112
    A[2,1] = alpha112(V)
    A[1,1] -= alpha112(V)

    #   i=2, j=3 => A[1,2] = beta112
    A[1,2] = beta112(V)
    A[2,2] -= beta112(V)

    #-- IF <-> I1: alpha4 from IF->I1, beta4 from I1->IF
    #   i=4, j=3 => A[3,2] = alpha4
    A[3,2] = alpha4(V)
    A[2,2] -= alpha4(V)

    #   i=3, j=4 => A[2,3] = beta4
    A[2,3] = beta4(V)
    A[3,3] -= beta4(V)

    #-- I1 <-> I2: alpha5 from I1->I2, beta5 from I2->I1
    #   i=5, j=4 => A[4,3] = alpha5
    A[4,3] = alpha5(V)
    A[3,3] -= alpha5(V)

    #   i=4, j=5 => A[3,4] = beta5
    A[3,4] = beta5(V)
    A[4,4] -= beta5(V)

    #-- C2 <-> IC2: alpha11 from C2->IC2, beta11 from IC2->C2
    #   i=2, j=7 => A[1,6] = alpha11
    A[1,6] = alpha11(V)
    A[6,6] -= alpha11(V)

    #   i=7, j=2 => A[6,1] = beta11
    A[6,1] = beta11(V)
    A[1,1] -= beta11(V)

    #-- C2 <-> C1: alpha12 from C2->C1, beta12 from C1->C2
    #   i=8, j=7 => A[7,6] = alpha12
    A[7,6] = alpha12(V)
    A[6,6] -= alpha12(V)

    #   i=7, j=8 => A[6,7] = beta12
    A[6,7] = beta12(V)
    A[7,7] -= beta12(V)

    #-- C1 <-> O: alpha13 from C1->O, beta13 from O->C1
    #   i=9, j=8 => A[8,7] = alpha13
    A[8,7] = alpha13(V)
    A[7,7] -= alpha13(V)

    #   i=8, j=9 => A[7,8] = beta13
    A[7,8] = beta13(V)
    A[8,8] -= beta13(V)

    #-- IF <-> O: alpha2 from IF->O, "beta33*beta2" from O->IF
    #   i=9, j=3 => A[8,2] = alpha2
    A[8,2] = alpha2(V)
    A[2,2] -= alpha2(V)

    #   i=3, j=9 => A[2,8] = beta33(V)*beta2(V)
    A[2,8] = beta33(V)*beta2(V)
    A[8,8] -= beta33(V)*beta2(V)

    return A

###############################################################################
# 3) Solve A(V) P_ss = 0, sum_i P_ss,i = 1  for the stationary distribution
###############################################################################
def get_stationary_probability(V):
    """
    Returns the 9-element steady-state probability vector P_ss(V).
    We do that by imposing sum(P_ss)=1 in the last row
    and A(V)*P_ss=0 in the other rows.
    """
    A = build_A_matrix(V).copy()
    b = np.zeros(9)

    # Overwrite last row with [1,1,1,1,1,1,1,1,1], set b[-1]=1
    b[-1] = 1.0
    A[-1,:] = 1.0

    P_ss = np.linalg.solve(A, b)
    return P_ss

def main():
    V_test = -15.0
    P = get_stationary_probability(V_test)
    print(f"Steady-state distribution at V={V_test} mV:\n", P)
    print("Sum of probabilities =", P.sum())
    # You can check that sum(P_ss)=1, and A(V)*P_ss ~ 0.

if __name__=="__main__":
    main()
```
